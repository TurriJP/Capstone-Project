```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

import maskslic as seg
```

```python
vv_src = rasterio.open('data/VV_VH.tif')
vv = vv_src.read()
vv = vv[:3, :, :]
vv = vv.transpose(1, 2, 0)
plt.imshow(vv)
```

```python
vv_slic = seg.slic(vv, n_segments=5000)
plt.imshow(vv_slic)
```
